```python
import requests

url = "https://api.tomorrow.io/v4/weather/forecast?location=TW6&timesteps=1d&apikey=kLyyt3sPWY6ERem3QAVEBwWGy1jS3BiR"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
forecast_airport_data = response.json()
forecast_airport_data
```

```python
forecast_airport_data['timelines']['daily']
```

```python
for forecast_data in forecast_airport_data['timelines']['daily']:
    print(forecast_data['time'])
```

```python
#Print the maximum temperature of each of the days.

for max_temp in forecast_airport_data['timelines']['daily']:
    print(max_temp['values']['temperatureApparentMax'])
```

```python
#10. Print only the day with the highest maximum temperature.
#New term: append. Append let's us add iterms to the list. 

highest_temp=[]

for max_temp in forecast_airport_data['timelines']['daily']:
   highest_temp.append(max_temp['values']['temperatureApparentMax'])

highest_temp_airport=max(highest_temp)

print("Highest max temp is:", highest_temp_airport)
```
